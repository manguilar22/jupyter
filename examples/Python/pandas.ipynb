```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("./../../datasets/census_data.csv")
df.head()
```

```python
df.dtypes
```

# Kind of Plots
* 'line' : line plot (default)
* 'bar' : vertical bar plot
* 'barh' : horizontal bar plot
* 'hist' : histogram
* 'box' : boxplot
* 'kde' : Kernel Density Estimation plot
* 'density' : same as 'kde'
* 'area' : area plot
* 'pie' : pie plot
* 'scatter' : scatter plot
* 'hexbin' : hexbin plot

```python
df[["age","education_num"]].plot(title="Line Plot",figsize=(12,3))
```

```python
df[["age"]].mean().plot.bar(figsize=(12,3))
```

```python
df[["age","capital_loss","capital_gain"]].mean().plot.box(figsize=(12,3))
```

```python
df[["capital_loss","capital_gain"]].plot(kind="hist",bins=20,cumulative=True)
```

```python
df.plot(kind="scatter",x="age",y="education_num",figsize=(12,3))
```
